Para la temperatura y entropía en una isoterma e isoentrópica. Solo se eligen valores

## Definir valores a gráficar

```python
from utils import saturated_diagram_ts, diagrama_ts, LABEL_SIZE
```

### Agua a vapor

```python
saturated_entropy, saturated_temperature = saturated_diagram_ts(0.0000, 9.1556)
```

```python
T_a = 250
T_b = T_a
T_c = 100
T_d = T_c
T = [
    [T_a, T_b],
    [T_b, T_c],
    [T_c, T_d],
    [T_d, T_a]
]
sat_t_f = saturated_temperature[:76]
sat_t_g = saturated_temperature[76:]
sat_s_f = saturated_entropy[:76]
sat_s_g = saturated_entropy[76:]
S_a = np.interp(T_a, sat_t_f, sat_s_f)
S_b = np.interp(T_b, sat_t_g[::-1], sat_s_g[::-1])
S_c = S_b
S_d = S_a
S = [
    [S_a, S_b],
    [S_b, S_c],
    [S_c, S_d],
    [S_d, S_a]
]
```

```python
fig, ax = diagrama_ts(S, T, "Ciclo de Carnot en rango agua vapor", True)
ax.plot(saturated_entropy, saturated_temperature, ":b", linewidth=2, alpha=0.8, label="Campana de\nSaturación")
ax.legend(fontsize=LABEL_SIZE)

fig.savefig("../Matplotlib/ciclo_carnot_ts_ideal")
```

### Vapor Saturado

```python
T_a = 400
T_b = T_a
T_c = 250
T_d = T_c
T = [
    [T_a, T_b],
    [T_b, T_c],
    [T_c, T_d],
    [T_d, T_a]
]

S_a = np.interp(T_d, sat_t_f, sat_s_f)
S_b = np.interp(T_c, sat_t_g[::-1], sat_s_g[::-1])
S_c = S_b
S_d = S_a
S = [
    [S_a, S_b],
    [S_b, S_c],
    [S_c, S_d],
    [S_d, S_a]
]
```
