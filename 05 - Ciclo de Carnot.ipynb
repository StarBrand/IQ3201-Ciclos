# Modelo Ciclo de Carnot

Importar librerias

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Definir funciones

Para la presión de un gas ideal isotérmico y adiabático

```python
# Valores dados para que calce el gráfico
R = 1
gamma = 10

def pressure_ideal(T, v):
    return R*T / v

def pressure_adiabatic(v, P1, v1):
    return P1*np.power(v1/v, gamma)
```

## Definir valores a gráficar

```python
Th = 850
Tc = 310

delta_v = 0.05
```

### Proceso Isotérmico

```python
vC = 2.5
vD = 2.0
```

Calculo de vA y vB

```python
PC = pressure_ideal(Tc, vC)
PD = pressure_ideal(Tc, vD)

vA = ((PD*(vD**gamma))/(R*Th))**(1/(gamma-1))
vB = ((PC*(vC**gamma))/(R*Th))**(1/(gamma-1))

v = np.linspace(vA-delta_v, vC+delta_v) # v1, 3
```

### Proceso adiabático

```python
v_2 = np.linspace(vB-delta_v, vC+delta_v)
v_4 = np.linspace(vA-delta_v, vD+delta_v)
```

```python
_, ax = plt.subplots(figsize=(8, 6))

ax.plot(v, pressure_ideal(Tc, v), '--b', label="Isoterma")
ax.plot(v, pressure_ideal(Th, v), '--b')
ax.plot(v_2, pressure_adiabatic(v_2, PC, vC), '-.r', label="Adiabática")
ax.plot(v_4, pressure_adiabatic(v_4, PD, vD), '-.r')
ax.set_title("Ciclo de Carnot\n", fontsize=16)
ax.set_xlabel("v", fontsize=14)
ax.set_ylabel("P", fontsize=14)
ax.tick_params(
    which='both',
    bottom=False,
    top=False,
    right=False,
    left=False,
    labelbottom=False,
    labelleft=False)
ax.grid()
ax.legend(fontsize=14)

plt.savefig("../Matplotlib/ciclo_carnot.png")
```
