# Eficiencia térmica

## Eficiencia Ciclo Otto

```python
import numpy as np
import matplotlib.pyplot as plt
from utils import LABEL_SIZE, TITLE_SIZE, FIGSIZE
```

```python
def eta(r, gamma):
    return 1 - (1 / np.power(r, gamma - 1))
```

### Motores de Gasolina

```python
_, ax = plt.subplots(figsize=FIGSIZE)

r = np.linspace(1, 14)
ax.plot(r, eta(r, 1.5), 'b', linewidth=3)
ax.fill_between([7, 10], 0.8, facecolor='r', alpha=0.3, label="Razones de Compresión\ntípicas de Gasolina")
ax.set_xlabel("Razón de Compresión, r", fontsize=LABEL_SIZE)
ax.set_ylabel("Eficiencia", fontsize=LABEL_SIZE)
ax.set_title("Eficiencia Otto", fontsize=TITLE_SIZE)
ax.set_xlim(0, 14)
ax.set_ylim(0, 0.8)
ax.legend(fontsize=LABEL_SIZE)
ax.grid()

plt.savefig("Matplotlib/ef_otto_gasolina.png")
```

### Motores a diferentes k

```python
_, ax = plt.subplots(figsize=FIGSIZE)

r = np.linspace(1, 14)
ax.plot(r, eta(r, 1.667), '--b', linewidth=3, label="k=1.667 Ar")
ax.plot(r, eta(r, 1.4), ':b', linewidth=3, label="k=1.4 aire")
ax.plot(r, eta(r, 1.3), '-.b', linewidth=3, label="k=1.3 CO2")
ax.set_xlabel("Razón de Compresión, r", fontsize=LABEL_SIZE)
ax.set_ylabel("Eficiencia", fontsize=LABEL_SIZE)
ax.set_title("Eficiencia Otto", fontsize=TITLE_SIZE)
ax.set_xlim(0, 14)
ax.set_ylim(0, 1.0)
ax.legend(fontsize=LABEL_SIZE)
ax.grid()

plt.savefig("Matplotlib/ef_otto_ks.png")
```

## Eficiencia Ciclo Diesel

```python
def eta_diesel(r, gamma, rc):
    return 1 - ((1 / np.power(r, gamma - 1))*((np.power(rc, gamma)-1)/(gamma*(rc - 1))))
```

### Motores Diesel

```python
_, ax = plt.subplots(figsize=FIGSIZE)

r = np.linspace(1, 24)
ax.plot(r, eta(r, 1.5), '-b', linewidth=3, label="rc=1 (Otto)")
ax.plot(r, eta_diesel(r, 1.5, 2), ':b', linewidth=3, label="2")
ax.plot(r, eta_diesel(r, 1.5, 3), '-.b', linewidth=3, label="3")
ax.plot(r, eta_diesel(r, 1.5, 4), '--b', linewidth=3, label="4")
ax.fill_between([12, 23], 0.8, facecolor='r', alpha=0.3, label="Razones de Compresión\ntípicas para motores Diesel")
ax.set_xlabel("Razón de Compresión, r", fontsize=LABEL_SIZE)
ax.set_ylabel("Eficiencia", fontsize=LABEL_SIZE)
ax.set_title("Eficiencia Diesel", fontsize=TITLE_SIZE)
ax.set_xlim(0, 24)
ax.set_ylim(0, 0.8)
ax.legend(fontsize=LABEL_SIZE)
ax.grid()

plt.savefig("Matplotlib/ef_diesel.png")
```

## Eficiencia Ideal Brayton

```python
def eta_brayton(rp, gamma):
    return 1 - (1 / np.power(rp, (gamma - 1) / gamma))
```

### Turbinas a Gas

```python
_, ax = plt.subplots(figsize=FIGSIZE)

rp = np.linspace(1, 25)
ax.plot(rp, eta_brayton(rp, 1.5), '-b', linewidth=3)
ax.fill_between([5, 20], 0.8, facecolor='r', alpha=0.3, label="Relaciones de Presiones\ntípicas para\nturbinas a gas")
ax.set_xlabel("Relación de Presiones, rp", fontsize=LABEL_SIZE)
ax.set_ylabel("Eficiencia", fontsize=LABEL_SIZE)
ax.set_title("Eficiencia Brayton", fontsize=TITLE_SIZE)
ax.set_xlim(0, 25)
ax.set_ylim(0, 0.7)
ax.legend(fontsize=LABEL_SIZE)
ax.grid()

plt.savefig("Matplotlib/ef_brayton_turbinas.png")
```

### Regeneración

```python
def eta_brayton_regeneración(rp, gamma, diff_t):
    return 1 - (diff_t*np.power(rp, (gamma - 1) / gamma))
```

```python
_, ax = plt.subplots(figsize=FIGSIZE)

rp = np.linspace(1, 25)
ax.plot(rp, eta_brayton(rp, 1.5), '-b', linewidth=3, label="Sin regeneración")
ax.plot(rp, eta_brayton_regeneración(rp, 1.5, 0.2), '--r', linewidth=3, label="Con Regeneración\nT1/T3=0.2")
ax.plot(rp, eta_brayton_regeneración(rp, 1.5, 0.25), '--k', linewidth=3, label="T1/T3=0.25")
ax.plot(rp, eta_brayton_regeneración(rp, 1.5, 0.33), '--g', linewidth=3, label="T1/T3=0.33")
ax.set_xlabel("Relación de Presiones, rp", fontsize=LABEL_SIZE)
ax.set_ylabel("Eficiencia", fontsize=LABEL_SIZE)
ax.set_title("Eficiencia Brayton con Regeneración", fontsize=TITLE_SIZE)
ax.set_xlim(0, 25)
ax.set_ylim(0, 0.8)
ax.legend(fontsize=LABEL_SIZE)
ax.grid()

plt.savefig("Matplotlib/ef_brayton_reg.png")
```
